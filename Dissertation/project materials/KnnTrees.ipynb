```python
import pandas as pd
from scipy.spatial import cKDTree
import matplotlib.pyplot as plt
import math
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
df2012AC1 = pd.read_csv("TreeLists_2012\TreeLists_2012\Aberfoyle_Centre1_Point Cloud Segmentation.csv")
df2012AC2 = pd.read_csv("TreeLists_2012\TreeLists_2012\Aberfoyle_Centre2_Point Cloud Segmentation.csv")
df2012AE = pd.read_csv("TreeLists_2012\TreeLists_2012\Aberfoyle_East_Normalize by DEM_Point Cloud Segmentation.csv")
df2012ASE = pd.read_csv("TreeLists_2012\TreeLists_2012\Aberfoyle_SE_Normalize by DEM_Point Cloud Segmentation.csv")
df2012ASW = pd.read_csv("TreeLists_2012\TreeLists_2012\Aberfoyle_SW_Normalize by DEM_Point Cloud Segmentation.csv")
df2012AW2 = pd.read_csv("TreeLists_2012/TreeLists_2012/Aberfoyle_West2_Normalize by DEM_Point Cloud Segmentation.csv")

df2021AV = pd.read_csv("TreeLists_2021\TreeLists_2021\Aber_Village_Normalize by DEM_Point Cloud Segmentation.csv")
df2021LW = pd.read_csv("TreeLists_2021\TreeLists_2021\Loch Ard West_Normalize by DEM_Point Cloud Segmentation.csv")
df2021LE = pd.read_csv("TreeLists_2021\TreeLists_2021\LochArdEast_Normalize by DEM_Point Cloud Segmentation.csv")



df2012 = pd.concat([df2012AC1, df2012AC2, df2012AE, df2012ASE, df2012ASW, df2012AW2], ignore_index=True)

df2021 = pd.concat([df2021AV, df2021LW, df2021LE], ignore_index=True)

df2012 = df2012[(df2012["TreePosZ"] > 4) & (df2012["TreePosZ"] < 40) & (df2012["CrownDiameter(metre)"] < 10) & (df2012["CrownDiameter(metre)"] > 0.5)]
df2021 = df2021[(df2021["TreePosZ"] > 4) &(df2021["TreePosZ"] < 40) & (df2021["CrownDiameter(metre)"] < 10) & (df2021["CrownDiameter(metre)"] > 0.5)]
```

```python
matchedTrees = pd.read_csv("matchedCells2012.csv")
```

```python
def genRelativeNeighbour(centralTree, nTree):
    relativePosX = nTree["TreePosX"] - centralTree["TreePosX"]
    relativePosY = nTree["TreePosY"] - centralTree["TreePosY"]
    
    # Convert to polar coordinates
    r = math.sqrt(relativePosX**2 + relativePosY**2)
    theta = math.atan2(relativePosY, relativePosX)  # Angle in radians
    theta_degrees = math.degrees(theta)  # Convert to degrees

    if(theta_degrees < 0): theta_degrees += 360
    
    return r, theta_degrees, nTree["TreePosZ"], nTree["CrownDiameter(metre)"], nTree["CrownArea(square metre)"], nTree["CrownVolume(cubic metre)"]


def createcKD(df):
    xs = df["TreePosX"]
    ys = df["TreePosY"]

    cords = list(zip(xs,ys))

    return cKDTree(cords)

cKDTree2012 = createcKD(df2012)
```

```python



def getKNN(centralTree, k, maxDist):
    cord = (centralTree["TreePosX"], centralTree["TreePosY"])
    vec = []

    dists, neighbourIDs = cKDTree2012.query(cord, k+1, distance_upper_bound=maxDist)


    for i in range(k):
        if(dists[i+1] > maxDist):
            vec.extend((-999,-999,-999,-999,-999,-999)) # padding
            continue
        vec.extend(genRelativeNeighbour(centralTree, df2012.iloc[neighbourIDs[i+1]]))
    
    return vec

getKNN(df2012.iloc[56], 10, 5)
```

```python
baseColumns = ["neighbourDistance","neighbourAngle","neighbourHeight","neighbourCrownDia","neighbourCrownArea","neighbourCrownVolume"]
addedColumns = []

def makeTable(k, maxDist):
    
    data = []

    for i in range(k):
        for s in baseColumns:
            addedColumns.append(str(i)+ "_" + s)
            data.append([])
    
    for index, row in matchedTrees.iterrows():
        ds = getKNN(row, k, maxDist)

        c = 0
        
        for d in ds:
            data[c].append(d)
            c += 1

        if index % 10000 == 0:
            print(f"progress: {100 * index/len(matchedTrees)}%")
    
    c = 0
    for cName in addedColumns:
        matchedTrees[cName] = data[c]
        c += 1
    
            


bigK = 5 # 15
maxBound = 15 #30

makeTable(bigK,maxBound)
        
```

```python
scaler = StandardScaler()

i = 0
for c in addedColumns:
    if i % 6 == 1:
        matchedTrees[f"{c}Norm"] = (matchedTrees[c] / 180) - 1
    else:
        # Mask padded values (-999) and fit only on valid data
        valid_mask = matchedTrees[c] != -999
        scaler.fit(matchedTrees[c][valid_mask].values.reshape(-1, 1))
        
        # Transform only valid values
        transformed_values = matchedTrees[c].values.reshape(-1, 1)
        transformed_values[valid_mask] = scaler.transform(transformed_values[valid_mask])

        # Assign back while keeping -999 unchanged
        matchedTrees[f"{c}Norm"] = transformed_values.flatten()
    i += 1

matchedTrees["TreePosXNorm"] = (((matchedTrees["TreePosX"] - matchedTrees["TreePosX"].min()) / (matchedTrees["TreePosX"].max() - matchedTrees["TreePosX"].min())) -0.5) * 2
matchedTrees["TreePosYNorm"] = (((matchedTrees["TreePosY"] - matchedTrees["TreePosY"].min()) / (matchedTrees["TreePosY"].max() - matchedTrees["TreePosY"].min())) -0.5) * 2
matchedTrees["Canopy CoverageNorm"] = (((matchedTrees["Canopy Coverage"] - matchedTrees["Canopy Coverage"].min()) / (matchedTrees["Canopy Coverage"].max() - matchedTrees["Canopy Coverage"].min())) -0.5) * 2

matchedTrees[["TreePosZNorm",
               "CrownDiameter(metre)Norm",
               "CrownDiameter(S-N)(metre)Norm",
               "CrownDiameter(E-W)(metre)Norm",
               "CrownArea(square metre)Norm",
               "CrownVolume(cubic metre)Norm", 
               "LAINorm",
               "Top HeightNorm",
               "AgeNorm",
               "YCNorm",
               "meanDBHNorm",
               "Basal AreaNorm",
               "Vol HaNorm"]] = scaler.fit_transform(matchedTrees[["TreePosZ",
                                                                    "CrownDiameter(metre)", 
                                                                    "CrownDiameter(S-N)(metre)",
                                                                    "CrownDiameter(E-W)(metre)",
                                                                    "CrownArea(square metre)",
                                                                    "CrownVolume(cubic metre)", 
                                                                    "LAI",
                                                                    "Top Height",
                                                                    "Age",
                                                                    "YC",
                                                                    "meanDBH",
                                                                    "Basal Area",
                                                                    "Vol Ha"]])


matchedTrees.to_csv(f"knn{bigK}-{maxBound}Trees.csv")
```
